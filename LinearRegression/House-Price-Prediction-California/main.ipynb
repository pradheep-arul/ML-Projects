```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Reading the data from the CSV file - Downloaded from https://www.kaggle.com/datasets/camnugent/california-housing-prices
original_data = pd.read_csv('data/housing.csv')
original_data
```

```python
# Info Step
# Getting the information of the data
original_data.info()
```

```python
# Remove the null values
data = original_data.dropna()
data.info()
```

```python
# Setting the input(X) and output(y) values
X = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

```python
# Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Merging the input and output values to find the Correlation between the input features and output
train_data = X_train.join(y_train)
train_data
```

```python
# Creating a correlation map to see the correlation between the input features and output('median_house_value')
correlation_map = train_data.corr(numeric_only=True)
correlation_map
```

```python
# Heatmap of the correlation data
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_map, annot=True, cmap='YlGnBu')
```

```python
# Looking at the input ranges
train_data.hist(figsize=(15, 8))
```

```python
# Preprocessing the data
# Normalizing the data - Removing the skewness
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
train_data.hist(figsize=(15, 8))
```

```python
# Split the 'ocean_proximity' column into multiple columns with boolean values of each category
# Converting from categorical to numerical data
ocean_proximity_split_columns = pd.get_dummies(train_data.ocean_proximity)
```

```python
# Drop the 'ocean_proximity' column and replace it with the new split columns
train_data = train_data.join(ocean_proximity_split_columns).drop(['ocean_proximity'], axis=1)
train_data
```

```python
# Plotting the Heatmap of the correlation data after splitting the 'ocean_proximity' column
plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='YlGnBu')
```

```python
# Plotting the scatter plot of 'latitude' and 'longitude' with 'median_house_value' - to see the geographical distribution of the data
# Most of the expensive houses are located near the coast - assuming the immaginary coast
plt.figure(figsize=(15, 8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='coolwarm')
```

```python
# Feature Engineering
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
# Plotting the Heatmap of the correlation data after Feature Engineering
plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(numeric_only=True), annot=True, cmap='YlGnBu')
```

```python
# Model LinearRegression Building
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Linear Regression Model
reg = LinearRegression()
scaler = StandardScaler()

# Splitting the data into input(X) and output(y) from the training data
X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg.fit(X_train_s, y_train)
```

```python
# Preparing the test data
test_data = X_test.join(y_test)

# Normalizing the test data
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

# Split the 'ocean_proximity' column into multiple columns with boolean values of each category
ocean_proximity_split_columns = pd.get_dummies(test_data.ocean_proximity)
# Drop the 'ocean_proximity' column and replace it with the new split columns
test_data = test_data.join(ocean_proximity_split_columns).drop(['ocean_proximity'], axis=1)


# Feature Engineering
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']

X_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
# Scale the X_test data
X_test_s = scaler.transform(X_test)
```

```python
# Score the model
reg.score(X_test_s, y_test)
```

Above result it not bad but we are gonna optimize the model to improve the result

```python
# Try the RandomForestRegressor model and see if it performs better
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train_s, y_train)
```

```python
forest.score(X_test_s, y_test)
```

RandomForestRegressor is performing better than the LinearRegression.

```python
# lets see if we could futher improve by optimising the Parameters using the GridSearchCV
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [3, 10, 30],
    'max_features': [2, 4, 6, 8],
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                            scoring='neg_mean_squared_error',
                            return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
# Peek at the best parameters
best_forest = grid_search.best_estimator_
best_forest
```

```python
# Score the best model
best_forest.score(X_test_s, y_test)
```

Above result was worse than the original score
Because the default n_estimators is 100 where we used a max of 30 for tuning.
So we should go more in that direction.

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [100, 200, 300],
    'min_samples_split': [2, 4],
    'max_depth': [None, 4, 8]

}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                            scoring='neg_mean_squared_error',
                            return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
# Peek at the best parameters
best_forest = grid_search.best_estimator_
best_forest
```

```python
# Score the best model
best_forest.score(X_test_s, y_test)
```

Default RandomForestRegressor was slightly better Compared toRandomForestRegressor with GridSearchCV
